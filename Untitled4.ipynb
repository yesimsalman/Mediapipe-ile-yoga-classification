```python
def detectPose(image, pose, display=True):
    
    
    # Create a copy of the input image.
    output_image = image.copy()
    
    # Convert the image from BGR into RGB format.
    imageRGB = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    
    # Perform the Pose Detection.
    results = pose.process(imageRGB)
    
    # Retrieve the height and width of the input image.
    height, width, _ = image.shape
    
    # Initialize a list to store the detected landmarks.
    landmarks = []
    
    # Check if any landmarks are detected.
    if results.pose_landmarks:
    
        # Draw Pose landmarks on the output image.
        mp_drawing.draw_landmarks(image=output_image, landmark_list=results.pose_landmarks,
                                  connections=mp_pose.POSE_CONNECTIONS)
        
        # Iterate over the detected landmarks.
        for landmark in results.pose_landmarks.landmark:
            
            # Append the landmark into the list.
            landmarks.append((int(landmark.x * width), int(landmark.y * height),
                                  (landmark.z * width)))
    
    # Check if the original input image and the resultant image are specified to be displayed.
    if display:
    
        # Display the original input image and the resultant image.
        plt.figure(figsize=[22,22])
        plt.subplot(121);plt.imshow(image[:,:,::-1]);plt.title("Original Image");plt.axis('off');
        plt.subplot(122);plt.imshow(output_image[:,:,::-1]);plt.title("Output Image");plt.axis('off');
        
        # Also Plot the Pose landmarks in 3D.
        mp_drawing.plot_landmarks(results.pose_world_landmarks, mp_pose.POSE_CONNECTIONS)
        
    # Otherwise
    else:
        
        # Return the output image and the found landmarks.
        return output_image, landmarks
```

```python
# Video için Poz işlevini ayarlayın.
pose_video = mp_pose.Pose(static_image_mode=False, min_detection_confidence=0.5, model_complexity=1)
 

 
# Diskte saklanan bir videodan okumak için VideoCapture nesnesini başlatın.
video = cv2.VideoCapture(r'C:\Users\salma\OneDrive\Masaüstü\yogagirl.mp4')
 
 
# Önceki çerçevenin zamanını saklamak için bir değişken başlatın.
time1 = 0
 

# Videoya başarıyla erişilene kadar yineleyin.
while video.isOpened():
    
    # Bir çerçeve okuyun.
    ok, frame = video.read()
    
    # Çerçevenin düzgün okunup okunmadığını kontrol edin.
    if not ok:
        
        
        break
    
   # Doğal (selfie görünümü) görselleştirme için çerçeveyi yatay olarak çevirin.
    frame = cv2.flip(frame, 1)
    
   # Çerçevenin genişliğini ve yüksekliğini alın
    frame_height, frame_width, _ =  frame.shape
    
    # En boy oranını korurken çerçeveyi yeniden boyutlandırın.
    frame = cv2.resize(frame, (int(frame_width * (640 / frame_height)), 640))
    
    
# Poz yer işareti tespiti gerçekleştirin.
    frame, _ = detectPose(frame, pose_video, display=False)
    
    # Bu çerçeve için zamanı şimdiki zamana ayarlayın.
    time2 = time()
    
    # Önceki ve bu kare süresi arasındaki farkın &gt; Sıfıra bölmeyi önlemek için 0.
    if (time2 - time1) > 0:
    
        
# Saniyedeki kare sayısını hesaplayın.
        frames_per_second = 1.0 / (time2 - time1)
        
       # Saniyede hesaplanan çerçeve sayısını çerçeveye yazın.
        cv2.putText(frame, 'FPS: {}'.format(int(frames_per_second)), (10, 30),cv2.FONT_HERSHEY_PLAIN, 2, (0, 255, 0), 3)
    
    
# Önceki kare süresini bu kare süresine güncelleyin.
    # Çünkü bu çerçeve bir sonraki yinelemede önceki çerçeve olacaktır.
    time1 = time2
    
   # Çerçeveyi görüntüleyin.
    cv2.imshow('Pose Detection', frame)
    
    
# Bir tuşa basılana kadar bekleyin.
    # Basılan tuşun ASCII kodunu alın
    k = cv2.waitKey(1) < 0xFF
    
   # 'ESC' basılıp basılmadığını kontrol edin.
    if(k == 27):
        
        
        break
 

# VideoCapture nesnesini bırakın.
video.release()
 
# Pencereleri kapat.
cv2.destroyAllWindows()
```

```python
def calculateAngle(landmark1, landmark2, landmark3):
   
 
   # Gerekli yer koordinatlarını alın.
    x1, y1, _ = landmark1
    x2, y2, _ = landmark2
    x3, y3, _ = landmark3
 
    # Üç nokta arasındaki açıyı hesapla
    angle = math.degrees(math.atan2(y3 - y2, x3 - x2) - math.atan2(y1 - y2, x1 - x2))
    
    # Açının sıfırdan küçük olup olmadığını kontrol edin.
    if angle < 0:
 
        
        angle += 360
    
    # Hesaplanan açıyı döndür
    return angle
```

```python
def classifyPose(landmarks, output_image, display=False):
    
    
    # Pozun etiketini başlatın. Bu aşamada bilinmiyor.
    label = 'Unknown Pose'
 
    # Etiketin görüntü üzerine yazılacağı rengi (Kırmızı) belirtin.
    color = (0, 0, 255)
    
    #Gerekli açıları hesaplayın.
    #----------------------------------------------------------------------------------------------------------------
    
    # Sol omuz, dirsek ve bilek noktaları arasındaki açıyı alın 
    left_elbow_angle = calculateAngle(landmarks[mp_pose.PoseLandmark.LEFT_SHOULDER.value],
                                      landmarks[mp_pose.PoseLandmark.LEFT_ELBOW.value],
                                      landmarks[mp_pose.PoseLandmark.LEFT_WRIST.value])
    
    # Sağ omuz, dirsek ve bilek noktaları arasındaki açıyı alın.
    right_elbow_angle = calculateAngle(landmarks[mp_pose.PoseLandmark.RIGHT_SHOULDER.value],
                                       landmarks[mp_pose.PoseLandmark.RIGHT_ELBOW.value],
                                       landmarks[mp_pose.PoseLandmark.RIGHT_WRIST.value])   
    
    # Sol dirsek, omuz ve kalça noktaları arasındaki açıyı alın. 
    left_shoulder_angle = calculateAngle(landmarks[mp_pose.PoseLandmark.LEFT_ELBOW.value],
                                         landmarks[mp_pose.PoseLandmark.LEFT_SHOULDER.value],
                                         landmarks[mp_pose.PoseLandmark.LEFT_HIP.value])
 
    # Sağ kalça, omuz ve dirsek noktaları arasındaki açıyı alın. 
    right_shoulder_angle = calculateAngle(landmarks[mp_pose.PoseLandmark.RIGHT_HIP.value],
                                          landmarks[mp_pose.PoseLandmark.RIGHT_SHOULDER.value],
                                          landmarks[mp_pose.PoseLandmark.RIGHT_ELBOW.value])
 
   # Sol kalça, diz ve ayak bileği noktaları arasındaki açıyı alın.
    left_knee_angle = calculateAngle(landmarks[mp_pose.PoseLandmark.LEFT_HIP.value],
                                     landmarks[mp_pose.PoseLandmark.LEFT_KNEE.value],
                                     landmarks[mp_pose.PoseLandmark.LEFT_ANKLE.value])
 
    # Sağ kalça, diz ve ayak bileği noktaları arasındaki açıyı alın
    right_knee_angle = calculateAngle(landmarks[mp_pose.PoseLandmark.RIGHT_HIP.value],
                                      landmarks[mp_pose.PoseLandmark.RIGHT_KNEE.value],
                                      landmarks[mp_pose.PoseLandmark.RIGHT_ANKLE.value])
    
    #----------------------------------------------------------------------------------------------------------------
    
   # Savaşçı II pozu mu yoksa T pozu mu olduğunu kontrol edin.
    # Her ikisinde de kollar düz ve omuzlar belirli açıda olmalıdır.
    #------------------------------------------------------------------------------------------------------------------
    
    # Her iki kolun da düz olup olmadığını kontrol edin.
    if left_elbow_angle > 165 and left_elbow_angle < 195 and right_elbow_angle > 165 and right_elbow_angle < 195:
 
        # Omuzların gereken açıda olup olmadığını kontrol edin.
        if left_shoulder_angle > 80 and left_shoulder_angle < 110 and right_shoulder_angle > 80 and right_shoulder_angle < 110:
 
   # Savaşçı II pozu olup olmadığını kontrol edin.
    #----------------------------------------------------------------------------------------------------------------
 
           # Bir bacağın düz olup olmadığını kontrol edin.
            if left_knee_angle > 165 and left_knee_angle < 195 or right_knee_angle > 165 and right_knee_angle < 195:
 
                # Diğer bacağın gerekli açıda bükülüp bükülmediğini kontrol edin.
                if left_knee_angle > 90 and left_knee_angle < 120 or right_knee_angle > 90 and right_knee_angle < 120:
 
                   # Warrior II pozu olan pozun etiketini belirtin.
                    label = 'Savasci II Pozu' 
                        
    #----------------------------------------------------------------------------------------------------------------
    
    #T pozu olup olmadığını kontrol edin.
    #----------------------------------------------------------------------------------------------------------------
    
            # Her iki bacağın da düz olup olmadığını kontrol edin
            if left_knee_angle > 160 and left_knee_angle < 195 and right_knee_angle > 160 and right_knee_angle < 195:
 
                # Ağaç pozu olan pozun etiketini belirtin.
                label = 'T Pozu'
 
    #----------------------------------------------------------------------------------------------------------------
    
    # Ağaç pozu olup olmadığını kontrol edin.
    #----------------------------------------------------------------------------------------------------------------
    
    # Bir bacak düz mü kontrol et
    if left_knee_angle > 165 and left_knee_angle < 195 or right_knee_angle > 165 and right_knee_angle < 195:
 
       # Diğer bacağın gerekli açıda bükülüp bükülmediğini kontrol edin.
        if left_knee_angle > 315 and left_knee_angle < 335 or right_knee_angle > 25 and right_knee_angle < 45:
 
          
# Ağaç pozu olan pozun etiketini belirtin.
            label = 'Agac Durusu'
                
    #----------------------------------------------------------------------------------------------------------------
    
   # Pozun başarıyla sınıflandırılıp sınıflandırılmadığını kontrol edin
    if label != 'Bilinmeyen Poz':
        
        # Etiketin görüntüye yazılacağı rengi (yeşile) güncelleyin.
        color = (0, 255, 0)  
    
    # Çıktı görüntüsüne etiketi yazın.
    cv2.putText(output_image, label, (10, 30),cv2.FONT_HERSHEY_PLAIN, 2, color, 2)
    
    
# Ortaya çıkan görüntünün görüntülenmek üzere belirtilip belirtilmediğini kontrol edin.
    if display:
    
        # Ortaya çıkan görüntüyü görüntüleyin.
        plt.figure(figsize=[10,10])
        plt.imshow(output_image[:,:,::-1]);plt.title("Output Image");plt.axis('off');
        
    else:
        
        # Çıktı görüntüsünü ve sınıflandırılmış etiketi döndürün.
        return output_image, label
```
